```python
# Set up the environment
from pathlib import Path
from utils import load_env, create_agent, stream_response, AgentOptions
load_env()

ARTIFACTS_DIR = Path("../../../2-Understanding_the_workflow/podcast-agent-artifacts")

def load_artifact(role: str) -> str:
    path = ARTIFACTS_DIR / f"{role}.txt"
    if not path.exists():
        raise FileNotFoundError(f"No artifact found for '{role}'. Run the artifact builder in Step 1 first.")
    return path.read_text()
```

# Meet your podcast studio team

In Step 1 you built five AI agents, each with a specific role in your podcast production pipeline. Their instructions were generated from your podcast concept and your chosen hosts.

This notebook lets you chat with each one so you can understand what they do, how they respond, and how their personality and role shapes their output — before you wire them together into a full workflow.

Your five agents are:

| Agent | Role |
|---|---|
| **Producer** | Sets the creative direction — angles, hooks, talking points |
| **Research** | Gathers facts, finds the story, surfaces what's interesting |
| **Script Writer** | Turns research into a host-ready script |
| **Editor** | Reviews the script for quality, flow, and factual accuracy |
| **Publisher** | Writes show notes, titles, and social copy |

---
## 1. The Producer

The Producer sets the creative direction for each episode. Give it a broad topic and it returns a sharp angle, a title, a one-sentence hook, and structured talking points — with notes on which host should lead each one.

Try asking it to plan an episode on any aspect of your podcast concept.

```python
producer = create_agent(AgentOptions(
    name="Producer",
    instructions=load_artifact("producer"),
))

await stream_response(producer.run(
    "",
    stream=True,
))
```

## 2. The Research Agent

The Research Agent digs into a topic and returns structured findings — key facts, open questions, and source suggestions. It knows what the show has and hasn't revealed, and flags gaps where the hosts should be careful not to overstate.

Try asking it to brief you on any theme, character, or mystery in your podcast.

```python
researcher = create_agent(AgentOptions(
    name="Research Agent",
    instructions=load_artifact("research"),
))

await stream_response(researcher.run(
    "",
    stream=True,
))
```

## 3. The Script Writer

The Script Writer takes research and talking points and turns them into a host-ready script — complete with banter, transitions, and each host's distinct voice. Notice how it writes differently for each host based on their personality.

Try giving it a topic or the output from the Producer above.

```python
script_writer = create_agent(AgentOptions(
    name="Script Writer",
    instructions=load_artifact("script-writer"),
))

await stream_response(script_writer.run(
    "",
    stream=True,
))
```

---
## 4. The Editor

The Editor reviews a draft script and returns specific, actionable notes — flagging pacing issues, factual overreach, moments where a host sounds out of character, and anything that should be cut or tightened.

Paste in a script excerpt and see what it catches.

```python
editor = create_agent(AgentOptions(
    name="Editor",
    instructions=load_artifact("editor"),
))

sample_script = """
KEN: So the polar bear — here's the thing — it's actually pretty simple. The DHARMA Initiative 
     brought animals to the island for experiments. Bear gets out. End of story.

MAYA: Okay but like, how did it survive? It's a tropical island. Polar bears need cold.

KEN: Animals are surprisingly adaptable. It probably just... figured it out.

MAYA: That feels like a big hand-wave. The show never actually explains it, right? 
      We're just assuming DHARMA did it.

KEN: Well, everyone assumes that. It's basically confirmed.
"""

await stream_response(editor.run(
    f"Please review this script excerpt and give me your editorial notes:\n\n{sample_script}",
    stream=True,
))
```

---
## 5. The Publisher

The Publisher takes a finished episode and produces everything needed to release it — a title, show notes, chapter markers, and social media copy. It writes for discoverability as much as for existing listeners.

Give it an episode title and topic and see what it produces.

```python
publisher = create_agent(AgentOptions(
    name="Publisher",
    instructions=load_artifact("publisher"),
))

await stream_response(publisher.run(
    """Episode title: "The Polar Bear Problem"
    
    Summary: Ken and Maya dig into one of Lost's most iconic mysteries — how a polar bear ended up 
    on a tropical island. They cover what the show actually confirms, what DHARMA files reveal later, 
    and why this small detail is a masterclass in the show's approach to world-building.
    
    Please write the show notes and a social media post for this episode.""",
    stream=True,
))
```
