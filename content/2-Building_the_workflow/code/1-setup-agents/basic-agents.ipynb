# Step 2 - Give your agent tools

```python
def get_meryl_streep_age () -> str:
    return "Meryl Streep is ageless."
```

```python
agent_with_no_tools = create_agent(
    AgentOptions(
        name="Celebrity Gossip Agent",
        instructions="You are my assistant. Answer my questions about celebrities and pop culture gossip",
    )
)

no_tools_response = await stream_response(agent_with_no_tools.run("How old is Meryl Streep?", stream=True))
```

```python
agent_with_tools = create_agent(
    AgentOptions(
        name="Celebrity Gossip Agent",
        instructions="You are my assistant. Answer my questions about celebrities and pop culture gossip",
        tools=[get_meryl_streep_age]
    )
)

tools_response = await stream_response(agent_with_tools.run("How old is Meryl Streep?", stream=True))
```
